```python
position_links = []

for position in ['qb', 'wr', 'rb', 'te']:

    stats_2019 = f'https://www.fantasypros.com/nfl/stats/{position}.php'

    position_links.append(stats_2019)

print(len(position_links))
```

```python
position_tables = []

for link in position_links:
    #read table from url
    table = pd.read_html(link, header=1)
    position_tables.append(table)

print(len(position_tables))
```

```python
cleaned_tables = []

for table in position_tables:
    table = table[0]
    cleaned_tables.append(table)
```

```python
# Clean RB Data

rb_2019 = cleaned_tables[1]

rb_2019 = rb_2019.rename(columns={"YDS.1": "REC_YDS", 'TD.1': 'REC_TDS'})

rb_2019
```

```python
# Clean WR Data

wr_2019 = cleaned_tables[2]

wr_2019 = wr_2019.rename(columns={"YDS.1": "RUSH_YDS", 'TD.1': 'RUSH_TDS'})

wr_2019
```

```python
# cleaned_dfs.append([qb_2019, rb_2019, wr_2019, te_2019])
# team_lists = []

for df in [qb_2019, rb_2019, wr_2019, te_2019]:
    #Split out Player Name and store in a temporary dataframe
    temp_df = df['Player'].str.split('(', n=1, expand=True)

    #add first name and last name to original salary dataframe
    df['Team_Name'] = temp_df[1]
    
    df['Team_Name'] = df['Team_Name'].str.strip(')')

    df['Player'] = df['Player'].str.split('(').str[0]
    
    df['Player'] = df['Player'].str.rstrip()
```

```python
# URls to pull data from
sos_table = 'https://www.fantasypros.com/nfl/strength-of-schedule.php'

#read table from url
table = pd.read_html(sos_table, header=0)
table
```

```python
sos_df = table[0]
```

```python
sos_df = sos_df[['Team', 'QB', 'RB', 'WR', 'TE']]

for col in ['QB', 'RB', 'WR', 'TE']:
    sos_df[col] = sos_df[col].astype(str).str.split().str[3]
```

```python
qb_sos = sos_df[['Team', 'QB']]
rb_sos = sos_df[['Team', 'RB']]
wr_sos = sos_df[['Team', 'WR']]
te_sos = sos_df[['Team', 'TE']]

qb_sos = qb_sos.rename(columns={'Team': 'Team_Name'})
rb_sos = rb_sos.rename(columns={'Team': 'Team_Name'})
wr_sos = wr_sos.rename(columns={'Team': 'Team_Name'})
te_sos = te_sos.rename(columns={'Team': 'Team_Name'})

qb_sos
```

```python
qb_data = qb_2019.merge(qb_sos, on='Team_Name')
rb_data = rb_2019.merge(rb_sos, on='Team_Name')
wr_data = wr_2019.merge(wr_sos, on='Team_Name')
te_data = te_2019.merge(te_sos, on='Team_Name')
```

```python
qb_data = qb_data.sort_values(by=['Rank'])
rb_data = rb_data.sort_values(by=['Rank'])
wr_data = wr_data.sort_values(by=['Rank'])
te_data = te_data.sort_values(by=['Rank'])
```

```python
qb_data = qb_data.rename(columns={'QB': 'SoS'})
rb_data = rb_data.rename(columns={'RB': 'SoS'})
wr_data = wr_data.rename(columns={'WR': 'SoS'})
te_data = te_data.rename(columns={'TE': 'Sos'})
```
