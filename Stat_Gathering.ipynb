```python
# URls to pull data from
qb_2019_stats = 'https://www.fantasypros.com/nfl/stats/qb.php'

#read table from url
table = pd.read_html(qb_2019_stats, header=1)
table
qb_2019 = table[0]

qb_2019 = qb_2019.rename(columns={"ATT.1": "RUSH_ATT", "YDS.1": "RUSH_YDS", 'TD.1': 'RUSH_TDS'})

qb_2019
```

```python
position_links = []

for position in ['qb', 'wr', 'rb', 'te']:

    stats_2019 = f'https://www.fantasypros.com/nfl/stats/{position}.php'

    position_links.append(stats_2019)

print(len(position_links))
```

```python
position_tables = []

for link in position_links:
    #read table from url
    table = pd.read_html(link, header=1)
    position_tables.append(table)

print(len(position_tables))
```

```python
cleaned_tables = []

for table in position_tables:
    table = table[0]
    cleaned_tables.append(table)
```

```python
# Clean QB Data

qb_2019 = cleaned_tables[0]

qb_2019 = qb_2019.rename(columns={"ATT.1": "RUSH_ATT", "YDS.1": "RUSH_YDS", 'TD.1': 'RUSH_TDS'})

qb_2019
```

```python
# Clean RB Data

rb_2019 = cleaned_tables[1]

rb_2019 = rb_2019.rename(columns={"YDS.1": "REC_YDS", 'TD.1': 'REC_TDS'})

rb_2019
```

```python
# Clean WR Data

wr_2019 = cleaned_tables[2]

wr_2019 = wr_2019.rename(columns={"YDS.1": "RUSH_YDS", 'TD.1': 'RUSH_TDS'})

wr_2019
```

```python
# Clean TE Data

te_2019 = cleaned_tables[3]

te_2019 = te_2019.rename(columns={"YDS.1": "RUSH_YDS", 'TD.1': 'RUSH_TDS'})

te_2019
```
